```python
import numpy as np
import matplotlib.pyplot as plt 
from calculator import Methods
```

```python
from calculator import Calculator
```

```python
cal=Calculator()
cal.read_excel('Data.xlsx')
order=2
M_p=cal.M_p[order]
cal.me.con.set_M_p(M_p)
ST,L_s=cal.ST[order],cal.L_init[order]
```

```python
fig=plt.figure(figsize=(10,20))
ax1=fig.add_subplot(211)
cal.me.cal_ML_simple(ST,L_s)
for i in range(len(cal.me.r)):
    if abs(cal.me.r[i]-cal.me.RCB)<1e-4:
        RCB_index=i 
        break 
r=cal.me.r
P=cal.me.P
T=cal.me.T
ax1.loglog(r,P)
ax1.loglog(r,T)
ax1.scatter(cal.me.RCB,cal.me.P[RCB_index],marker='o')
ax1.scatter(cal.me.RCB,cal.me.T[RCB_index],marker='o')

ax2=fig.add_subplot(212)
M=cal.me.M
ax2.loglog(r,M)
ax2.scatter(cal.me.RCB,cal.me.M[RCB_index],marker='o')

plt.show()
```
