```python
import numpy as np
from calculator import Methods,Calculator
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
    me=Methods()
    #print(me.con.P_0)

    p=np.linspace(100,1e+5,1000)
    t=np.linspace(10,1000,1000)
    P,T=np.meshgrid(p,t)
    stepP=1
    stepT=1e-3
    sigma_0=me.sigma(P,T)
    dP=(me.sigma(P+stepP,T)-sigma_0)/stepP/sigma_0
    dT=(me.sigma(P,T+stepT)-sigma_0)/stepT/sigma_0
```

#### 上面的两幅图分别是在P、T坐标下对$\mathrm{d} \ln \sigma / \mathrm{d} P$和$\mathrm{d} \ln \sigma / \mathrm{d} T$作图；下面两幅图则是分别取P,T坐标下（另一个坐标固定在第300个点）下对$\mathrm{d} \ln \sigma / \mathrm{d} P$和$\mathrm{d} \ln \sigma / \mathrm{d} T$作图

```python
    fig=plt.figure(dpi=300)
    ax1=fig.add_subplot(221,projection='3d')
    ax1.plot_surface(P,T,dP,cmap='rainbow')
    ax2=fig.add_subplot(222,projection='3d')
    ax2.plot_surface(P,T,dT,cmap='rainbow')
    
    ax3=fig.add_subplot(223)
    ax3.semilogx(P[300],dP[300])
    ax4=fig.add_subplot(224)
    ax4.semilogx(T[:,300],dT[:,300])
    plt.show()
```
