```python
import numpy as np 
import matplotlib.pyplot as plt
from calculator import Methods,Calculator
from scipy import optimize
```

```python
cal=Calculator()
cal.read_excel('Data2.xlsx')
order=12
L_s=cal.L_init[order]
ST=cal.ST[order]
```

```python
cal.me.cal_ML_simple(ST,L_s)
P1=cal.me.P
T1=cal.me.T
L1=cal.me.L
RCB1=cal.me.RCB_index
cal.me.cal_ML_simple_B(ST,L_s,True,1e-15,1e-8)
P2=cal.me.P
T2=cal.me.T
L2=cal.me.L
RCB2=cal.me.RCB_index
r=cal.me.r
```

```python
fig=plt.figure(dpi=400)
ax1=fig.add_subplot(221)
ax1.plot(r,P1,label='r-P(without B)')
ax1.plot(r,P2,label='r-P(with B)')
ax1.scatter(r[RCB1],P1[RCB1],marker='o')
ax1.scatter(r[RCB2],P2[RCB2],marker='o')
plt.legend(loc='upper right')
ax2=fig.add_subplot(222)
ax2.plot(r,T1,label='r-T(without B)')
ax2.plot(r,T2,label='r-T(with B)')
ax2.scatter(r[RCB1],T1[RCB1],marker='o')
ax2.scatter(r[RCB2],T2[RCB2],marker='o')
plt.legend(loc='upper right')
ax3=fig.add_subplot(223)
ax3.loglog(P1,T1,label='P-T(without B)')
ax3.loglog(P2,T2,label='P-T(with B)')
ax3.scatter(P1[RCB1],T1[RCB1],marker='o')
ax3.scatter(P2[RCB2],T2[RCB2],marker='o')
plt.legend(loc='upper right')
plt.plot()
```
