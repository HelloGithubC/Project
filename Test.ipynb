```python
fig=plt.figure(dpi=200)
ax1=fig.add_subplot(121)
ax1.semilogx(dG,L_in,label=r'$ST-L_{in}$')
ax2=fig.add_subplot(122)
ax2.plot(dG,g_in,label=r'$ST-g_{in}$')
plt.legend(loc='upper right')
plt.show()
```
