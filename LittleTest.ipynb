```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def ode_fun(Y,x):
    return [Y[1],6*Y[0]/x**2+4*Y[1]/x]
```

```python
dG=1e-11
initial=[0.0,dG]
x=np.linspace(1,0.001,5000)
result=odeint(ode_fun,initial,x)
g=result[:,0]
```

```python
fig=plt.figure(dpi=100)
plt.plot(x,g,'b--')
#plt.plot(x,g/(1e-11*(x**3-x**(-2))),'b--')
plt.plot(x,dG/7*(x**6-x**(-1)))
plt.xlim([0.001,0.02])
plt.show()
```
